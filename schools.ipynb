```python
%matplotlib inline

from os.path import dirname, join
import csv
import matplotlib.pyplot as plt
DATA_DIR = join('data', 'schools', 'extracted')
frpm_dataname = join(DATA_DIR, 'frpm-2014.csv')
sat_dataname = join(DATA_DIR, 'sat-2014.csv')
```

```python
with open(frpm_dataname, 'r') as rf:
    frpm_data = list(csv.DictReader(rf))
    
with open(sat_dataname, 'r') as rf:
    sat_data = list(csv.DictReader(rf))
```

```python
sat_schools = [d for d in sat_data if d['school_name'] and int(d['number_of_test_takers']) > 20]
```

```python
usable_cds = [d['cds'] for d in sat_schools]
frpm_schools = [d for d in frpm_data if d['cds'] in usable_cds]

```

```python
plt.scatter([d['adjusted_pct_eligible_frpm_k12'] for d in frpm_schools], 
            [d['percent_scores_gte_1500'] for d in sat_schools])
```

```python
plt.scatter([d['avg_math_score'] for d in sat_schools], 
            [d['avg_reading_score'] for d in sat_schools])
```

```python
plt.scatter([d['avg_writing_score'] for d in sat_schools], 
            [d['avg_reading_score'] for d in sat_schools])
```
